```python
# Fill missing 'country' with mode (most common country)
df['country'] = df['country'].fillna(df['country'].mode().values[0])
print(df['country'].isna().sum())
```

```python
# Fill missing 'rating' with mode
df['rating'] = df['rating'].fillna(df['rating'].mode().values[0])
print(df['rating'].isna().sum())
```

```python
# Check Missing Values after imputation
print("\nMissing Values After Cleaning")
print(df.isnull().sum())
```

```python
# Fix inconsistent text formatting - uppercase 'type'
df['type'] = df['type'].str.upper()
df.head()
```

## Part-A: Data Normalization

```python
# 1. Min-Max Normalization on release_year
df['release_year_MinMax'] = (df['release_year'] - df['release_year'].min()) / (df['release_year'].max() - df['release_year'].min())
print("\nMin-Max Normalized release_year")
print(df[['title', 'release_year', 'release_year_MinMax']].head(10))
```

```python
# 2. Z-score Normalization on release_year
df['release_year_Zscore'] = (df['release_year'] - df['release_year'].mean()) / df['release_year'].std()
print("\nZ-score Normalized release_year")
print(df[['title', 'release_year', 'release_year_Zscore']].head(10))
```

```python
# 3. Decimal Scaling on release_year
df['release_year_Decimal'] = df['release_year'] / 10000
print("\nDecimal Scaled release_year")
print(df[['title', 'release_year', 'release_year_Decimal']].head(10))
```

```python
# Normalize Multiple Columns (Min-Max)
cols = ['release_year']
df_norm = (df[cols] - df[cols].min()) / (df[cols].max() - df[cols].min())
print("\nNormalized Dataset")
print(df_norm.head(10))
```

## NLP Techniques on Text Data in Python

```python
# Download required NLTK packages
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
import nltk
# Tokenization - Word Tokenization
from nltk.tokenize import word_tokenize
text = df['description'].iloc[0] # Taking the first description from the DataFrame
tokens = word_tokenize(text)
print(tokens)
```

```python
# Sentence Tokenization
from nltk.tokenize import sent_tokenize
# Use a longer text (first 5 descriptions combined)
long_text = " ".join(df['description'].dropna().head(5).tolist())
sentences = sent_tokenize(long_text)
print(sentences)
```

```python
# Stop Word Removal
# Stop words such as is, the, and, in are removed
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
words = word_tokenize(text)
filtered_words = [word for word in words if word.lower() not in stop_words]
print(filtered_words)
```

```python
# Stemming - reduces words to their root form
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
# Use unique keywords from netflix descriptions
words_to_stem = ["Documentaries", "streaming", "Movies", "watching", "Studios", "Filming"]
for w in words_to_stem:
    print(w, ":", stemmer.stem(w))
```

```python
# Lemmatization - converts words to their dictionary form
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
netflix_words = ["movies", "series", "episodes", "running", "directed", "starring"]
for w in netflix_words:
    print(w, ":", lemmatizer.lemmatize(w))
```

```python
# Part of Speech (POS) Tagging
# POS tagging identifies grammatical roles of words
from nltk import pos_tag
words = word_tokenize(text)
pos_tags = pos_tag(words)
print(pos_tags)
```

```python
'''
Meaning of POS Tags:
NNP  - Proper Noun (singular)
NN   - Noun (singular)
PRP  - Personal Pronoun
VBP  - Verb (non-3rd person singular present)
VBG  - Verb (gerund / present participle)
JJ   - Adjective
IN   - Preposition or subordinating conjunction
DT   - Determiner
.    - Punctuation
'''
```

```python
# Word Frequency Count on Netflix descriptions
from nltk.probability import FreqDist
# Combine all descriptions and analyse
all_descriptions = " ".join(df['description'].dropna().tolist())
all_words = word_tokenize(all_descriptions)
filtered_all = [word.lower() for word in all_words if word.lower() not in stop_words and word.isalpha()]
fdist = FreqDist(filtered_all)
print("Top 20 most common words in Netflix descriptions:")
print(fdist.most_common(20))
```

# Basic Charts and Visual Encoding

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid", palette="muted")
plt.rcParams.update({'figure.dpi': 120, 'axes.titlesize': 14,
                     'axes.labelsize': 12, 'xtick.labelsize': 10,
                     'ytick.labelsize': 10})

# ── Load & Prepare ───────────────────────────────────────────
df = pd.read_csv('/content/netflix_titles.csv')

year_df = df['release_year'].value_counts().sort_index().reset_index()
year_df.columns = ['Release_Year', 'Count']
year_df = year_df[year_df['Release_Year'].between(2012, 2021)]

type_year = df.groupby(['release_year','type']).size().unstack(fill_value=0).reset_index()
type_year = type_year[type_year['release_year'].between(2012, 2021)]

rating_df = df['rating'].value_counts().head(8).reset_index()
rating_df.columns = ['Rating', 'Count']

movies = df[df['type']=='Movie'].copy()
movies['duration_min'] = movies['duration'].str.extract(r'(\d+)').astype(float)
movies = movies.dropna(subset=['duration_min'])

sample = df.sample(400, random_state=42).copy()
sample['duration_min'] = sample['duration'].str.extract(r'(\d+)').astype(float)
sample = sample.dropna(subset=['duration_min'])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid", palette="muted")
plt.rcParams.update({'figure.dpi': 120, 'axes.titlesize': 14,
                     'axes.labelsize': 12, 'xtick.labelsize': 10,
                     'ytick.labelsize': 10})

# ── Load & Prepare ───────────────────────────────────────────
df = pd.read_csv('netflix_titles.csv')

year_df = df['release_year'].value_counts().sort_index().reset_index()
year_df.columns = ['Release_Year', 'Count']
year_df = year_df[year_df['Release_Year'].between(2012, 2021)]

type_df   = df['type'].value_counts().reset_index()
type_df.columns = ['Type', 'Count']

rating_df = df['rating'].value_counts().head(6).reset_index()
rating_df.columns = ['Rating', 'Count']

top_countries = df['country'].value_counts().head(10).reset_index()
top_countries.columns = ['Country', 'Total']
top_countries['Movies']   = top_countries['Country'].apply(
    lambda c: ((df['country']==c) & (df['type']=='Movie')).sum())
top_countries['TV_Shows'] = top_countries['Country'].apply(
    lambda c: ((df['country']==c) & (df['type']=='TV Show')).sum())

print(top_countries.head())
```

> ## More Visual Analysis of Netflix Content

```python
# Initialization

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid", palette="muted")
plt.rcParams.update({'figure.dpi': 120, 'axes.titlesize': 14,
                     'axes.labelsize': 12, 'xtick.labelsize': 10,
                     'ytick.labelsize': 10})

# ── Load Dataset ─────────────────────────────────────────────
data = pd.read_csv('/content/netflix_titles.csv')

# ── Remove Redundant Columns ─────────────────────────────────
data = data.drop(['show_id', 'description'], axis=1)
data.head()

# ── Info of the Dataset ──────────────────────────────────────
data.info()

# ── Fix Data Types ───────────────────────────────────────────
data['date_added']   = pd.to_datetime(data['date_added'].str.strip(), errors='coerce')
data['release_year'] = data['release_year'].astype('int64')
data.info()

# ── Recheck ──────────────────────────────────────────────────
data.head()

# ── Derived Column: Content Age ──────────────────────────────
data['content_age'] = 2024 - data['release_year']
data.iloc[50:100]

# ── Most Recent Date Added ───────────────────────────────────
data['date_added'].max()

# ── Titles Added in the Latest Year ─────────────────────────
latest_year = data['date_added'].dt.year.max()
latest_data = data[data['date_added'].dt.year == latest_year]
latest_data.head()

# ── Shape ────────────────────────────────────────────────────
latest_data.shape

# ── Country-wise Count ───────────────────────────────────────
data['country'].value_counts()

# ── Unique Countries ─────────────────────────────────────────
data['country'].unique()

# ── Number of Unique Countries ───────────────────────────────
data['country'].nunique()

# ── Country-wise Aggregated Table ────────────────────────────
country_type = data.groupby('country')['type'].value_counts().unstack(fill_value=0).reset_index()
country_type.columns.name = None
for col in ['Movie', 'TV Show']:
    if col not in country_type.columns:
        country_type[col] = 0
country_type['Total'] = country_type['Movie'] + country_type['TV Show']
country_type = country_type.rename(columns={'Movie':'Movies','TV Show':'TV_Shows'})
country_type

# ── India ────────────────────────────────────────────────────
country_type[country_type['country'] == 'India']

# ── United States ────────────────────────────────────────────
country_type[country_type['country'] == 'United States']
```
